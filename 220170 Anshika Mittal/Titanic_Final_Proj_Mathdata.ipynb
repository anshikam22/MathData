```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data=pd.read_csv("train.csv")
test_data=pd.read_csv("test.csv")
test_data_id=test_data["PassengerId"]
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
def clean_data(df):
  df=df.drop(["Ticket", "Cabin", "Name","PassengerId"], axis=1)

  cols=["SibSp", "Parch", "Fare", "Age"]
  for col in cols:
    df[col].fillna(df[col].mean(), inplace=True)

  df.Embarked.fillna("A",inplace=True)
  return df
```

```python
train_data = clean_data(train_data)
```

```python
train_data.head()
```

```python
test_data=clean_data(test_data)
```

```python
test_data.head()
```

```python
from sklearn import preprocessing
le= preprocessing.LabelEncoder()

columns = ["Sex","Embarked"]

for col in columns:
  train_data[col]=le.fit_transform(train_data[col])
  test_data[col]=le.fit_transform(test_data[col])
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
X=train_data.drop("Survived", axis=1)
y=train_data["Survived"]

X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
predictions=DecisionTreeClassifier(max_leaf_nodes=10,random_state=0)
predictions.fit(X_train,y_train)
```

```python
y_pred=predictions.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
confusion_matrix(y_test,y_pred)
```

```python
submission_pred=predictions.predict(test_data)
```

```python
df=pd.DataFrame({"PassengerId":test_data_id.values,
                 "Survived": submission_pred,})
```

```python
df.to_csv("my_submission.csv",index=False)
```
